```python
import matplotlib.pyplot as plt
import numpy as np

c = 299792.458
H0 = 73.4
Omega_m = 0.3

def H_z(z):
    return H0*np.sqrt(Omega_m*(1+z)**3+(1-Omega_m))

z = np.linspace(0,100)

plt.plot(z,c/H_z(z))
plt.xlabel("z")
plt.ylabel(r"$D_H$ Mpc")
plt.title(r"$D_H$ vs z")
plt.plot()
```
